## Representing one trial with plain variables
Everything an experiment records on a trial is just a value with a type.

````python
# one trial of a simple contrast-detection task, stored in plain variables
subject_id = 7              # an integer label for the participant
contrast = 0.08             # stimulus contrast on this trial (a float, 0-1)
stimulus_present = True     # was a signal actually shown? (a boolean)
responded_yes = True        # did the subject report "I saw it"? (a boolean)
reaction_time = 0.412       # seconds (a float)

# f-strings are the modern, readable way to build text out of variables
print(f"subject {subject_id} responded in {reaction_time * 1000:.0f} ms")
````

:::{admonition} Advanced tips
:class: tip
**f-strings** (`f"...{value}..."`, Python 3.6+) have replaced older `%` and
`.format()` styling for almost all new code. The `:.0f` inside the braces is a
format spec -- here, "show zero decimal places".
:::

Booleans are a special kind of integer and take a value of 0 (False) or 1 (True). This is often useful if you need to know how many times events are true because you perform basic arithmetic on them:

````python
# True behaves like 1 and False like 0 when you do arithmetic
print(stimulus_present + responded_yes)   # 2
````

## Comparing floating-point numbers...
Floats are stored in binary and cannot represent most decimal values exactly.

````python
# this surprises almost everyone the first time they see it
print(0.1 + 0.2)               # 0.30000000000000004, not 0.3
print( (0.1 + 0.2) == 0.3)     # False!
````

:::{admonition} Advanced tips: never test floats for exact equality
:class: tip
An AI assistant (or a tired human) might happily write `if reaction_time == 0.5:`
or `if contrast == 0.1:`. That code *runs without any error*, but the comparison
can silently be `False` because of binary rounding, so the branch never executes
and trials get misclassified. Use a tolerance instead:

```python
import math
math.isclose(reaction_time, 0.5, abs_tol=1e-9)   # more robust
```

:::

:::{admonition} Advanced tips
:class: tip
A light touch of **type hints** documents what a variable is meant to hold. They
do not change how the code runs, but they help readers (and AI tools) reason about
your intent:

```python
reaction_time: float = 0.412
subject_id: int = 7
```
:::
