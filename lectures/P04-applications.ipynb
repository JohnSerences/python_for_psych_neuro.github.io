# P04: Applications and advanced tips

Dictionaries, sets, and files are the workhorses of organizing experimental data:
grouping trials by condition, counting open-ended responses, tracking unique
stimulus IDs, and reading raw log files. The standard library has excellent tools
that make all of this short and clear.

## Grouping trials by condition with `defaultdict`

```python
from collections import Counter, defaultdict

# group reaction times by condition
rts_by_cond = defaultdict(list)
trials = [("congruent", 512), ("incongruent", 640),
          ("congruent", 488), ("incongruent", 705)]
for cond, rt in trials:
    rts_by_cond[cond].append(rt)     # no need to check if the key exists yet

print(dict(rts_by_cond))

# a dict comprehension gives the mean RT per condition
mean_by_cond = {c: sum(v) / len(v) for c, v in rts_by_cond.items()}
print(mean_by_cond)
```

## Counting open-ended responses with `Counter`

```python
responses = "red red blue green red blue".split()
print(Counter(responses))                  # how often each answer appeared
print(Counter(responses).most_common(1))   # the single most common answer
```

## Sets for unique stimulus IDs

```python
shown = {101, 102, 103, 102, 101}    # a set automatically drops duplicates
print(shown)
print(f"{len(shown)} unique stimuli shown")
```

## Reading a log file the modern way: `pathlib` + `with`

```python
from pathlib import Path

# pathlib handles file paths portably across macOS/Windows/Linux
log_path = Path("trial_log.txt")
log_path.write_text("trial,rt\n1,512\n2,640\n")

# 'with' guarantees the file is closed even if an error happens mid-read
with log_path.open() as f:
    for line in f:
        print(line.strip())

log_path.unlink()    # delete the temporary file
```

:::{admonition} Double check: two quiet assumptions about dicts and files
:class: warning
* **Dict ordering.** Since Python 3.7 dictionaries preserve *insertion* order --
  but that is not the same as *sorted* order. Code that loops over a dict assuming
  conditions come out alphabetically (or in any particular order) can mislabel
  results. Sort explicitly when order matters: `sorted(d.items())`.
* **Hand-parsing CSVs.** Splitting log lines on commas (`line.split(",")`) breaks
  the moment a field itself contains a comma, an unexpected encoding, or quoted
  text. It often "works" on your test file and corrupts the real one. Use `pandas`
  (P07) or the `csv` module for anything beyond a toy example.
:::

:class: tip

:::
